```python
import os
import re

# Function to fix the filename
def fix_filename(name):
    # Replace pattern like '50 MT_501.jpg' -> '50_MT_501.jpg'
    fixed_name = re.sub(r'^(\d+)\s+([A-Za-z])', r'\1_\2', name)
    return fixed_name
```

```python
folder = 'C:/Users/admin/Documents/GitHub/Pill_Identification/Pill_Jpeg_Processed'

for old_name in os.listdir(folder):
    if not old_name.lower().endswith('.jpg'):
        continue
    new_name = fix_filename(old_name)
    if new_name != old_name:
        os.rename(os.path.join(folder, old_name), os.path.join(folder, new_name))
        print(f"Renamed: {old_name} -> {new_name}")
```

```python
import cv2

def image_resize(image, width = None, height = None, inter = cv2.INTER_AREA):
    # initialize the dimensions of the image to be resized and
    # grab the image size
    dim = None
    (h, w) = image.shape[:2]

    # if both the width and height are None, then return the
    # original image
    if width is None and height is None:
        return image

    # check to see if the width is None
    if width is None:
        # calculate the ratio of the height and construct the
        # dimensions
        r = height / float(h)
        dim = (int(w * r), height)

    # otherwise, the height is None
    else:
        # calculate the ratio of the width and construct the
        # dimensions
        r = width / float(w)
        dim = (width, int(h * r))

    # resize the image
    resized = cv2.resize(image, dim, interpolation = inter)

    # return the resized image
    return resized
```

```python
import cv2
import numpy as np
import os
import shutil

# Input and output base directories
input_root = "C:/Users/admin/Documents/GitHub/Pill_Identification/Pill_Jpeg_2025"
output_root = "C:/Users/admin/Documents/GitHub/Pill_Identification/Pill_Jpeg_Processed"

# Desired output size
target_size = 500

# Remove the output folder if it exists
# if os.path.exists(output_root):
#     shutil.rmtree(output_root)

# Supported image extensions
image_extensions = ('.jpg', '.jpeg', '.png', '.bmp')

used_filenames = set()

for root, dirs, files in os.walk(input_root):
    for file in files:
        if file.lower().endswith(image_extensions):
            input_path = os.path.join(root, file)
            filename = os.path.basename(file)

            # Check for duplicate filename
            if filename in used_filenames:
                print(f"⚠️ Skipped duplicate: {filename}")
                continue
            used_filenames.add(filename)

            output_path = os.path.join(output_root, filename)

            # Load and resize image
            image = cv2.imread(input_path)
            if image is None:
                print(f"❌ Failed to load: {input_path}")
                continue

            resized = image_resize(image, width=target_size, height=target_size)

            # Save image
            cv2.imwrite(output_path, resized)
            print(f"✅ Saved: {output_path}")
```
